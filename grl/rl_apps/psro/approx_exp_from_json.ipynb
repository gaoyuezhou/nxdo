```python
import json
import matplotlib.pyplot as plt
with open("approx_exploitability_3_seeds.json", "r+") as jsonFile:
    data = json.load(jsonFile)
```

```python
data
```

```python
all_logs = {}
iters = [1,   500,  1000,  1500,  2000,  2500,  3000,  4000,  5000, 6000,  7000,  8500, 10500, 13000, 15500]
for seed in data.keys():
    approx_exp = [[], []]
    for player in [0, 1]:
        dct = data[seed][f'player_{player}']['br_policy_reward_mean']
        for iter in iters:
            approx_exp[player].append(dct[f'iter_{iter}'])
    all_logs[seed] = approx_exp
```

```python
cnt = 1
for seed in all_logs.keys():
    print(seed)
    game_exp = all_logs[seed]
    for player_idx in range(len(game_exp)):
        player_exp = game_exp[player_idx]
        plt.plot(iters, player_exp, label=f'game {seed} player {player_idx}')
    plt.legend(loc='upper right')
    plt.savefig(f"approx_exp_seed{cnt}")
    plt.show()
    plt.close()
    cnt += 1
```
